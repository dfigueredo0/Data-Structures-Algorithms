```python
# Code to create a list of all alphabets
import random
from string import ascii_lowercase

alpha = []

for x in ascii_lowercase:
    alpha.append(x)
        
def random_string(length):
    string_length = random.randint(1, length)
    random_string = ''.join(random.choice(alpha) for _ in range(string_length))
    return random_string
```

```python
class CustomHeap:
    def _parent(idx):
        return (idx - 1) // 2
    
    def _left(idx):
        return idx*2 + 1

    def _right(idx):
        return idx*2 + 2

    def __init__(self, iterable=None):
        if not iterable:
            self.data = []
        else:
            self.data = list(iterable)
            last_internal_idx = CustomHeap._parent(len(self.data)-1)
            for i in range(last_internal_idx, -1, -1):
                self._heapify(i)

    def add(self, x):
        self.data.append(x)
        idx = len(self.data) - 1
        while idx > 0:
            pidx = CustomHeap._parent(idx)
            if self.data[pidx] < self.data[idx]:
                self.data[pidx], self.data[idx] = self.data[idx], self.data[pidx]
                idx = pidx
            else:
                break    
                                
    def max(self):
        assert len(self) > 0  # if condition returns True, then nothing happens, else, AssertionError is raised
        return self.data[0]

    def _heapify(self, idx):
        while idx < len(self.data):
            lidx = CustomHeap._left(idx)
            ridx = CustomHeap._right(idx)
            minidx = idx
            if lidx < len(self.data) and len(self.data[lidx]) < len(self.data[idx]):
                minidx = lidx
            if ridx < len(self.data) and len(self.data[ridx]) < len(self.data[minidx]):
                minidx = ridx
            if minidx != idx:
                self.data[idx], self.data[minidx] = self.data[minidx], self.data[idx]
                idx = minidx
            else:
                break

    def pop_max(self):
        assert len(self) > 0
        ret = self.data[0]
        self.data[0] = self.data[-1]
        del self.data[-1]
        self._heapify(0)
        return ret        
    
    def __bool__(self):
        return len(self.data) > 0

    def __len__(self):
        return len(self.data)

    def __repr__(self):
        return repr(self.data)


def main():
    custom_heap = CustomHeap()
    max_length = int(input("Enter the maximum length of random string: "))
    num_nodes = int(input("Enter the number of nodes in the heap: "))

    for _ in range(num_nodes):
        random_str = random_string(max_length)
        custom_heap.add(random_str)

    # Display the final heap
    print("The final heap is")
    print(custom_heap)

if __name__ == "__main__":
    main()
```
