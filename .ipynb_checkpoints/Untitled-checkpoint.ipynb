```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Constants
L = 1.1  # Pendulum length in meters
r = 0.3  # Wheel radius in meters
t_w = 0.04  # Wheel thickness in meters
m_left = m_right = 0.5  # Wheel mass in kg
m_b = 5.0  # Segway body mass in kg
m_p = 1.0  # Pendulum mass in kg
r_p = 0.02  # Pendulum radius in meters
g = 9.81  # Acceleration due to gravity in m/s^2

# Moment of inertia
I_p = (1/3) * m_p * L**2  # Pendulum
I_w = 0.5 * (m_left + m_right) * (r**2 + t**2)  # Wheel

# Initial conditions
theta_0 = 0.0  # Initial angle in radians
omega_0 = 0.0  # Initial angular velocity in radians/s
phi_0 = 0.0  # Initial lateral position
v_0 = 0.0  # Initial lateral velocity

# Differential drive model equations
def model(y, t):
    theta, omega, phi, v = y

    # Equations of motion for forward motion
    dtheta_dt = omega
    domega_dt = (m_p * g * L * np.sin(theta) - 0.5 * m_p * L * omega**2 * np.sin(2 * theta)
                  - m_p * r * omega**2 * np.cos(theta) - 0.5 * (m_left + m_right) * r * g * np.sin(phi)) / (0.5 * I_p + m_p * r**2)

    # Equations of motion for lateral movement
    dphi_dt = v
    dv_dt = (m_left * r * omega**2 + m_right * r * omega**2 - m_left * m_right * r**2 * omega**2 * np.sin(phi)
             + m_p * r * omega**2 * np.cos(phi) - m_p * L * omega**2 * np.sin(theta) * np.cos(phi) - (m_left + m_right) * g * np.sin(phi)) / (m_left + m_right + m_p)

    return [dtheta_dt, domega_dt, dphi_dt, dv_dt]

# Time points
t = np.linspace(0, 10, 1000)

# Initial conditions vector
y0 = [theta_0, omega_0, phi_0, v_0]

# Solve ODE
solution = odeint(model, y0, t)

# Extracting results
theta = solution[:, 0]
phi = solution[:, 2]

# Plotting
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.plot(t, theta, label='Pendulum Angle')
plt.title('Forward Motion')
plt.xlabel('Time (s)')
plt.ylabel('Angle (rad)')
plt.legend()

plt.subplot(122)
plt.plot(t, phi, label='Lateral Position')
plt.title('Lateral Movement')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Define Values
t = np.linspace(0,10,100)
time = np.ones_like(t)
Length = 1.10 #Length of Pendulum, meters
Radius = 0.30 #Wheel Radius, meters
Thick_W = 0.04 #Wheel Thickness, meters
Mass_m = 0.5 #Wheel Mass for L and R, Kg
Mass_S = 5 #Mass of Segway, kg
Mass_P = 1 #Mass of Pendulum, kg
Radius_P = 0.02 #Pendulum Radius, meters
Theta = np.cos(time) #Pendulum Oscillation
Theta_Der = -np.sin(time) #Pendulum Oscillation Derivative
Angular_Vel_L =  np.ones_like(t) #radians^-1
Angular_Vel_R = Angular_Vel_L * 2


#Kinetic Energy of Pendulum
Omega_1 = np.array([0, 0, Theta_Der], dtype=object)
Omega_1_T = np.transpose(Omega_1)
Inertia_B = [[(1/12)*Mass_P*(3*(Radius_P**2)+(Length**2)),0,0],[0,(1/12)*Mass_P*(3*(Radius_P**2)+(Length**2)),0],[0,0,0.5*Mass_P*Radius_P**2]]
Completed_Matrix = np.matmul(Omega_1_T,Inertia_B)
Completed_Matrix_1 = np.matmul(Completed_Matrix,Omega_1)
Kinetic_P = 0.5*Mass_P*((0.5*Length*Theta_Der)**2)+0.5*Completed_Matrix_1

plt.plot(t, Kinetic_P)
plt.grid(True)
plt.plot()


#Kinetic Energy of Wheel (Left)
Omega_2 = np.array([0 ,0 ,Angular_Vel_L], dtype=object)
Omega_2_T = np.transpose(Omega_2)
Inertia_A = [[(1/12)*Mass_m*(3*(Radius**2)+(Thick_W**2)),0,0],[0,(1/12)*Mass_m*(3*(Radius**2)+(Thick_W**2)),0],[0,0,0.5*Mass_m*Radius**2]]
Completed_Matrix2 = np.matmul(Omega_2_T,Inertia_A)
Completed_Matrix_2 = np.matmul(Completed_Matrix2,Omega_2)
Kinetic_Wheel_L = 0.5*Mass_m*((Angular_Vel_L*Radius)**2)+0.5*Completed_Matrix_2

plt.plot(t, Kinetic_Wheel_L)
plt.grid(True)
plt.plot()


#Kinetic Energy of Right and Left Wheel
Omega_2 = np.array([0, 0, Angular_Vel_R], dtype=object)
Omega_2_T = np.transpose(Omega_2)
Inertia_A = [[(1/12)*Mass_m*(3*(Radius**2)+(Thick_W**2)),0,0],[0,(1/12)*Mass_m*(3*(Radius**2)+(Thick_W**2)),0],[0,0,0.5*Mass_m*Radius**2]]
Completed_Matrix2 = np.matmul(Omega_2_T,Inertia_A)
Completed_Matrix_2 = np.matmul(Completed_Matrix2,Omega_2)
Kinetic_Wheel_R = 0.5*Mass_m*((Angular_Vel_R*Radius)**2)+0.5*Completed_Matrix_2

plt.plot(t, Kinetic_Wheel_L+Kinetic_Wheel_R)
plt.grid(True)
plt.plot()
```
