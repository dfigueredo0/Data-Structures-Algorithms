```python
import timeit
import matplotlib.pyplot as plt

def fibonacci_iter(n):
    fibonacci = [] 
    a, b = 0, 1

    fibonacci.append(a)

    for _ in range(1, n):
        a, b = b, a + b
        fibonacci.append(a)

    return fibonacci

def fibonacci_rec(n):
    fibonacci = [] 

    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        fibonacci = fibonacci_rec(n - 1)
        fibonacci.append(fibonacci[-1] + fibonacci[-2])
        
    return fibonacci
    
def main():
    n = int(input("Enter number for fibonacci sequence range"))
    print(f"Fibonacci sequence of {n} terms: {fibonacci_iter(n)}")
    print(f"Fibonacci sequence of {n} terms: {fibonacci_rec(n)}")

    exe_time = []
    
    for i in range(n + 1):
        time_taken = timeit.timeit(lambda: fibonacci_rec(i), number=10)
        exe_time.append(time_taken)
    
    plt.plot(range(n + 1), exe_time, marker='o')
    plt.xlabel('Input Size (N)')
    plt.ylabel('Time (seconds)')
    plt.title('Execution Time of fibonacci_rec() for Input Sizes')
    plt.grid(True)
    
    plt.show()
    
main()
```
