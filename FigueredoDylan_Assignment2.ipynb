```python
import pandas as pd
import random

d1 = {'C1': [random.randint(0, 100) for _ in range(3)],
         'C2': [random.randint(0, 100) for _ in range(3)],
         'C3': [random.randint(0, 100) for _ in range(3)]}
df1 = pd.DataFrame(data = d1, index=['r1', 'r2', 'r3'])

d2 = {'C1': [random.randint(0, 100) for _ in range(3)],
         'C2': [random.randint(0, 100) for _ in range(3)],
         'C3': [random.randint(0, 100) for _ in range(3)]}
df2 = pd.DataFrame(data =d2, index=['r4', 'r5', 'r6'])

dfC = pd.concat([df1, df2])

dfC.plot(kind = 'line')

d3 = {'C4': [random.randint(0, 100) for _ in range(3)],
         'C5': [random.randint(0, 100) for _ in range(3)],
         'C6': [random.randint(0, 100) for _ in range(3)]}
df3 = pd.DataFrame(data = d3, index=['r2', 'r3', 'r4'])

dfC2 = pd.concat([df1, df3], axis=1, join='inner')

print(df1, '\n')
print(df2, '\n')
print(dfC, '\n')
print(df3, '\n')
print(dfC2, '\n')
```
